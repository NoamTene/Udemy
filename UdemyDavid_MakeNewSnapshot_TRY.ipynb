```python
# Return Costs
def add_costs( soup ):
    #print(f"In add_costs: {type(soup)}")
    
    regex = re.compile('^buy-box__element')
    resultSet = soup.find_all('div', attrs={'class': regex})                 # type(resultSet) = bs4.element.ResultSet
   
    #print(resultSet )
    
    ndx = 0
    empt_ndx = 0
    # the 'resultSet'     - type 'bs4.element.ResultSet'
    #   - each element of type 'bs4.element.Tag' (may or may not have 'text')
    
    
    # Check if "Free Course" ( is  element exusts:  <meta property="udemy_com:price" content="Free">
    description = soup.find('meta', attrs={'property':'udemy_com:price'}) 
    # if FOUND
    if (description):
    #entry.description = description.get('content')
        #print( description.get('content'), len(description.get('content')))
        if( description.get('content') == 'Free'):
            file1.write(f", 0, 0, 0 ")
            print(f"free, ", end="")
            return 0
    
    
    for i in resultSet:
        # print(ndx, type(i))
        if( i.text ==""  ):  continue
         
        found_ndx = (i.text).find("Current price",0, len(i.text))
        #print(f"Next item {len(i.text)}")
        #print(i.text )
        
        if ( found_ndx != -1 ):
            #print(ndx, found_ndx)
            found_txt = i.text[found_ndx: found_ndx+70]
            #print( found_txt )
 
            #print( found_txt )
            txt_list = found_txt.split(" ")
            #print(txt_list, found_txt)
            curr_price_lst1 = txt_list[2].split("\n")
            curr_price_lst2 = txt_list[4].split("\n")
            #print(f"lst2 {curr_price_lst1} lst4 {curr_price_lst2}")
            curr_price = curr_price_lst1[0][1:len(curr_price_lst1[0])] 
            orig_price = curr_price_lst2[0][1:len(curr_price_lst2[0])] 
            #print(curr_price_lst2[4], len(curr_price_lst2[4]) )
            discount = curr_price_lst2[4][0:len(curr_price_lst2[4])-1] 
            #print(float(curr_price), float(orig_price), int(discount))
            
            
            file1.write(f", {float(orig_price)}, {float(curr_price)}, {int(discount)} ")
            return 0                
                 
          
    file1.write(f", -1, -1, -1 ")            
    return -1
```
